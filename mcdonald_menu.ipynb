```python
import pandas as pd
df = pd.read_csv("McDonald_Menu.csv")
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
df.columns = df.columns.str.replace(" ", "_").str.lower()
```

```python
df
```

```python
df_cleaned = df.dropna() #dropped rows with missing values
```

```python
df_cleaned.isnull().sum()
```

```python
print(df_cleaned.duplicated())
```

```python
top_calories = df_cleaned.sort_values(by='calories', ascending=False).head(10)
print(top_calories[['item', 'calories']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
sns.barplot(x='calories', y='item', data=top_calories, palette='Reds_r')
plt.title("Top 10 Highest Calorie Items at McDonald's")
plt.xlabel("Calories")
plt.ylabel("Menu Item")
plt.show()
```

```python
avg_calories = df_cleaned.groupby('category')['calories'].mean().sort_values(ascending=False)
print(avg_calories)
```

```python
avg_calories.plot(kind='bar', figsize=(10,5), color='blue')
plt.title("Average Calories per Food Category")
plt.ylabel("Average Calories")
plt.xticks(rotation=45)
plt.show()
```

```python
df_cleaned['category'].value_counts()
```

```python
high_sodium = df_cleaned.sort_values(by='sodium', ascending=False).head(10)
print(high_sodium[['item', 'sodium']])
```

```python
sns.scatterplot(data=df_cleaned, x='total_fat', y='calories', hue='category')
plt.title("Calories vs Total Fat")
plt.xlabel("Total Fat (g)")
plt.ylabel("Calories")
plt.legend(loc='best')
plt.show()
```
